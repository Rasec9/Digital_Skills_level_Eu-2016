```python
#Importiamo pandas e numpy
import pandas as pd
import numpy as np
```

```python
#creiamo una lista che contenga le etichette che corrispondono ai quattro livelli di digital skills e il livello nullo
skill_levels = ['I_DSK_NA', 'I_DSK_X', 'I_DSK_L', 'I_DSK_B', 'I_DSK_AB']
```

```python
#chiedo attraverso .loc di localizzare le righe dove, nella colonna var_indic_is, sono presenti le etichette che ci servono
d_skills.loc[d_skills['var_indic_is'].isin(skill_levels)]
```

```python
#salvo una copia di backup del database completo
d_skills_backup = d_skills
```

```python
#salvo il database d'interesse come d_skills
d_skills = d_skills.loc[d_skills['var_indic_is'].isin(skill_levels)]
```

```python
#Adesso salvo come d_skills soltanto la parte della matrice riguardante tutti gli individui.
d_skills = d_skills.loc[d_skills['var_ind_type'] == 'IND_TOTAL']
```

```python
#Divido la matrice in due matrici: una con coloro che hanno usato internet negli ultimi 3 mesi, una con tutti.
d_skills_3_months = d_skills.loc[d_skills['var_unit'] == 'PC_IND_IU3']
d_skills_all = d_skills.loc[d_skills['var_unit'] == 'PC_IND']
```

```python
#Controllo che nella matrice a 3 mesi non ci siano quelli che non hanno usato internet negli ultimi 3 mesi
d_skills_3_months.loc[d_skills['var_indic_is'] == 'I_DSK_NA']
```

```python
#Creiamo una lista con le etichette relative ai paesi che ci servono
country_list = ['IT', 'EU28', 'EU15', 'PT', 'DE']
```

```python
#Creo una nuova matrice filtrata sulla base delle etichette in country_list. Conterrò solo i valori dei paesi di interesse
country_filter = d_skills_all.loc[d_skills_all['var_geo'].isin(country_list)]
country_filter
```

```python
year2016 = country_filter.loc[country_filter['variable'] == 2016]
```

```python
year2016
```

```python
#Pulisco la tabella in modo che sia di facile lettura
operational = year2016.filter(['var_indic_is','var_geo','value'], axis=1)
```

```python
operational
```

```python
#Riordino le colonne
operational = operational[['var_geo', 'var_indic_is', 'value']]
```

```python
operational
```

```python
#Faccio in modo che gli indici diventino le nazioni
operational.set_index('var_geo', inplace = True)
```

```python
#Rinomino le colonne
operational = operational.rename(columns={"var_indic_is": "livello abilità", "value": "%"})
```

```python
operational
```

```python
#Riordino la tabella e setto alcuni dei valori (il livello delle digital skill) come delle colonne e non più delle righe.
#In questo modo ottengo una matrice "standard"
operational.pivot(columns = 'livello abilità', values='%')
```

```python
#Cambio il nome delle variabili (che ora sono le nostre colonne). Rendo così la tabella leggibile.
#Rinonino di nuovo la tabella sempre operational, così che sostituisca la precedente.
#Perdo la visualizzazione a "tabella classica", no panic, basta richiamare il pivot
operational = operational.replace({'I_DSK_AB': "Above Basic", 'I_DSK_B': "Basic", 'I_DSK_L': "Low", 'I_DSK_X': "No skills", 'I_DSK_NA': "NA"})
```

```python
operational.pivot(columns = 'livello abilità', values='%')
```
